```python
#Uppgift d
import matplotlib.pyplot as plt
import random

for g in range(10):  #Här skapar vi våran loop
    x = random.uniform(-1,1)  # X, detta är våran första randomiserade nummer, våran x punkt
    y = random.uniform(-1,1)  # Y, detta är våran första randomiserade nummer, våran y punkt
    plt.plot(x,y,'ro') #Vi sätter nu in plottern i loppen så att den skapar 10 olike punkter i diagrammet, om vi satte denna utanför loppen så skulle den endast visa dom högsta värden på diagrammet.
plt.xlim(-1,1) #Vi vill visa våran diagram mellan punkterna -1,1
plt.ylim(-1,1)
plt.show() # Vi sätter plt.show utanför loppen för att den inte ska skapa 10 olika diagram, den skapar endast 1 diagram med 10 punkter.
```
