```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.arange(11)
def f(x):
    y = 3*x+1
    return y

plt.plot(x,f(x))
plt.title("f(x) = 3x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

```python
import numpy as np 
import matplotlib.pyplot as plt

t = np.linspace(0,6)
y = -4*t**2+24*t+1

plt.plot(t,y)
plt.ylabel("y")
plt.xlabel("t")
plt.show()
```
